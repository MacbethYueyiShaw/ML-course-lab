```python
data.loc[data.price_range <= 1, 'price_range'] = 0
data.loc[data.price_range >= 2, 'price_range'] = 1
y = data['price_range']
x = data.drop('price_range', axis = 1)
```

```python
labels = ["low cost",  "high cost"]
values = data['price_range'].value_counts().values
colors = ['yellow','lightblue']
fig1, ax1 = plt.subplots()
ax1.pie(values, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.set_title('after convet')
plt.show()
```

 - data normalization

```python
x = data.drop('price_range', axis=1)
x_min = data.drop('price_range', axis=1).min()
x_max = data.drop('price_range', axis=1).max()
x = (x - x_min) / (x_max - x_min)
x
```

- Split the original ‘train.csv’ into ‘train.csv’, ‘valid.csv’ and ‘test.csv’
- with the ratio of 0.8 : 0.1 : 0.1, respectively.

```python
x_train, x_tmp, y_train, y_tmp = train_test_split(x, y, test_size=0.2, random_state=101, stratify=y)
x_valid, x_test, y_valid, y_test = train_test_split(x_tmp, y_tmp, test_size=0.5, random_state=101, stratify=y_tmp)

print(x_train.shape)
print(x_valid.shape)
print(x_test.shape)
```

# 2.SVM

```python
svm_x_train = x_train.copy()
svm_y_train = y_train.copy()
svm_x_valid = x_valid.copy()
svm_y_valid = y_valid.copy()
svm_x_test = x_test.copy()
svm_y_test = y_test.copy()
svm = LinearSVC()
```

- training

```python
timer_start = time.perf_counter()
svm.fit(svm_x_train, svm_y_train)
timer_end = time.perf_counter()
svm_time_train = timer_end - timer_start

svm_time_train
```

- valid

```python
timer_start = time.perf_counter()
svm_y_predict = svm.predict(svm_x_valid)
timer_end = time.perf_counter()
svm_time_valid = timer_end - timer_start
print(svm_time_valid)
```

- valid Accuracy

```python
confusion_matrix = metrics.confusion_matrix(svm_y_valid, svm_y_predict)
accurate = metrics.accuracy_score(svm_y_valid, svm_y_predict)
accurate
```

- test

```python
timer_start = time.perf_counter()
svm_y_predict = svm.predict(svm_x_test)
timer_end = time.perf_counter()
svm_time_test = timer_end - timer_start
print(svm_time_test)
```

- Test Accuracy

```python
confusion_matrix = metrics.confusion_matrix(svm_y_test, svm_y_predict)
accurate = metrics.accuracy_score(svm_y_test, svm_y_predict)
accurate
```
